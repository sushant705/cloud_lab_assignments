```python
#Part I: Randomized Sales Data Generation (NumPy)
#1. Initialize Random Seed
#2. Generate Sales Data
#• Create a NumPy array of random sales data for 12 months (shape = (12, 4)), where each column represents a product category:
#• Electronics, Clothing, Home & Kitchen, Sports.
#• Sales valuesshould be between 1000 and 5000 units per month.
#• The range should be randomized using the seed.
#3. Convert to DataFrame
#• Convert the cleaned NumPy array into a Pandas DataFrame with:
#i. Columns: Electronics, Clothing, Home & Kitchen, Sports
#ii. Index: Monthsfrom Jan to Decimport numpy as np

import pandas as pd

roll_no = 86
np.random.seed(roll_no)
```

```python
sales = np.random.randint(1000, 5000, size=(12, 4))
```

```python
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

df = pd.DataFrame(sales, columns=["Electronics","Clothing","Home & Kitchen","Sports"], index=months)

print(df)
```

```python
#Part II: Data Manipulation and Analysis(Pandas)
#1. Display the first 5 rows and summary statistics of the DataFrame.
#2. Calculate the totalsales per category and totalsales per month.
#3. Calculate the average sales growth between consecutive monthsfor each category.
#4. Add the following columns to the DataFrame:
#• Total Sales: Sum of all categories per month.
#• Growth Rate: Percentage change in Total Salesfrom the previous month.
#5. Apply a randomized discount:
#• If your roll number is even, apply a 10% discount to the Electronics category.
#• If odd, apply a 15% discount to the Clothing category

print(df.head())
print(df.describe())
```

```python
print("Total sales per category:")
print(df.sum())

print("Total sales per month:")
print(df.sum(axis=1))
```

```python
growth = df.pct_change()
print(growth)
```

```python
df["Total Sales"] = df.sum(axis=1)
df["Growth Rate"] = df["Total Sales"].pct_change()

print(df)
```

```python
if roll_no % 2 == 0:
    df["Electronics"] = df["Electronics"] * 0.9
else:
    df["Clothing"] = df["Clothing"] * 0.85

print(df)
```

```python
#Part III: Visualizations(Matplotlib and Seaborn)
#1. Plot monthly salestrendsfor each category using line plots.
#2. Create the following plots: Box plots to show the sales distribution for each category.import matplotlib.pyplot as plt

df[["Electronics","Clothing","Home & Kitchen","Sports"]].plot()
plt.title("Monthly Sales")
plt.show()
```

```python
import seaborn as sns

sns.boxplot(data=df[["Electronics","Clothing","Home & Kitchen","Sports"]])
plt.show()
```

```python
# For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
# i. Find element-wise absolute value
# ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each row.
# iii. Mean, Median and Standard Deviation of flattened array, of each column, and each row

import numpy as np

array = np.array([[1, -2, 3],[-4, 5, -6]])

# Absolute
print(np.abs(array))

# Percentiles
print(np.percentile(array.flatten(), [25,50,75]))
print(np.percentile(array, [25,50,75], axis=0))
print(np.percentile(array, [25,50,75], axis=1))

# Mean, Median, Std
print(np.mean(array))
print(np.median(array))
print(np.std(array))

print(np.mean(array, axis=0))
print(np.mean(array, axis=1))
```

```python
# Write a Python program to swap two elementsin a set by converting it to a list first

#Original set
my_set = {10, 20, 30, 40}

# Convert set to list
my_list = list(my_set)

print("Original list:", my_list)

# Swap elements (example: swap index 0 and 2)
i, j = 0, 2

my_list[i], my_list[j] = my_list[j], my_list[i]

# Convert back to set
my_set = set(my_list)

print("After swapping:", my_set)
```
